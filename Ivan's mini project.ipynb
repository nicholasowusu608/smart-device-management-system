```python

"""
Description: A menu-driven system managing virtual smart home appliances,
demonstrating inheritance, encapsulation, validation, and polymorphism.
"""

class SmartDevice:
    """Parent class representing a generic smart home device."""

    def __init__(self, device_id: str, name: str):
        # Validation Check: Ensure device ID is not empty
        if not device_id or not device_id.strip():
            raise ValueError("Device ID cannot be empty.")

        # Private Attributes (Encapsulation)
        self.__device_id = device_id.strip()
        self.__power_status = False  # False indicates OFF, True indicates ON

        # Public Attribute
        self.Name = name

    # Getter for device_id property
    @property
    def device_id(self) -> str:
        return self.__device_id

    # Getter for power_status property
    @property
    def power_status(self) -> bool:
        return self.__power_status

    def turn_on(self):
        """Sets the power status to True."""
        self.__power_status = True
        print(f"[{self.Name}] has been turned ON.")

    def turn_off(self):
        """Sets the power status to False."""
        self.__power_status = False
        print(f"[{self.Name}] has been turned OFF.")

    def display_info(self):
        """Displays standard system metadata."""
        status = "ON" if self.__power_status else "OFF"
        print(f"Device Name: {self.Name}")
        print(f"Device ID  : {self.__device_id}")
        print(f"Power State: {status}")


class TemperatureSensor(SmartDevice):
    """Child class inheriting from SmartDevice with temperature tracking features."""

    def __init__(self, device_id: str, name: str, initial_temp: float = 24.0):
        # Initialize parent attributes via super()
        super().__init__(device_id, name)
        # Unique attribute
        self.temperature = initial_temp

    def read_temperature(self):
        """Outputs current sensory temperature if powered on."""
        if not self.power_status:
            print(f"Operation Failed: [{self.Name}] is currently powered OFF.")
            return
        print(f"[{self.Name}] Current Temperature Reading: {self.temperature}°C")

    def display_info(self):
        """Overrides parent display method to include unique capability data."""
        super().display_info()
        print(f"Capability : Temperature Monitoring ({self.temperature}°C)")


class SmartLight(SmartDevice):
    """Child class inheriting from SmartDevice with brightness level controls."""

    def __init__(self, device_id: str, name: str, initial_brightness: int = 50):
        super().__init__(device_id, name)
        # Enforcing boundaries during initialization
        if 0 <= initial_brightness <= 100:
            self.brightness = initial_brightness
        else:
            self.brightness = 50

    def increase_brightness(self, amount: int = 10):
        """Increases intensity while strictly capping value at 100."""
        if not self.power_status:
            print(f"Operation Failed: [{self.Name}] is currently powered OFF.")
            return
        self.brightness = min(100, self.brightness + amount)
        print(f"[{self.Name}] Brightness increased to {self.brightness}%.")

    def decrease_brightness(self, amount: int = 10):
        """Decreases intensity while strictly stopping at 0."""
        if not self.power_status:
            print(f"Operation Failed: [{self.Name}] is currently powered OFF.")
            return
        self.brightness = max(0, self.brightness - amount)
        print(f"[{self.Name}] Brightness decreased to {self.brightness}%.")

    def display_info(self):
        """Overrides parent display method to include unique capability data."""
        super().display_info()
        print(f"Capability : Luminescence adjustment ({self.brightness}%)")


class SecurityCamera(SmartDevice):
    """Child class inheriting from SmartDevice with continuous recording states."""

    def __init__(self, device_id: str, name: str):
        super().__init__(device_id, name)
        # Unique attribute
        self.recording_status = False

    def start_recording(self):
        """Activates capture feed if base system power is available."""
        if not self.power_status:
            print(f"Operation Failed: Cannot record. [{self.Name}] is powered OFF.")
            return
        self.recording_status = True
        print(f"[{self.Name}] Video surveillance feed capturing... [RECORDING START]")

    def stop_recording(self):
        """Deactivates live camera stream recording."""
        self.recording_status = False
        print(f"[{self.Name}] Video stream tracking halted. [RECORDING STOP]")

    def display_info(self):
        """Overrides parent display method to include unique capability data."""
        super().display_info()
        rec_label = "ACTIVE" if self.recording_status else "INACTIVE"
        print(f"Capability : Video Capture Processing Status: ({rec_label})")



# Menu-Driven Interface Console Framework

def main():
    # Instantiate the target required smart devices
    sensor = TemperatureSensor("TMP-001", "Living Room Thermostat", 22.5)
    light = SmartLight("LGT-002", "Ceiling Chandelier", 75)
    camera = SecurityCamera("CAM-003", "Front Porch Security Cam")

    # Store references systematically to manipulate polymorphically
    devices = {
        "1": sensor,
        "2": light,
        "3": camera
    }

    while True:
        print("\n" + "="*45)
        print("    SMART DEVICE MANAGEMENT CORE ENGINE    ")
        print("="*45)
        print("1. Display Device Information")
        print("2. Turn Device On")
        print("3. Turn Device Off")
        print("4. Read Temperature (Sensor Only)")
        print("5. Adjust Brightness (Smart Light Only)")
        print("6. Start/Stop Recording (Camera Only)")
        print("7. Exit Application")
        print("="*45)

        choice = input("Select an operational menu choice (1-7): ").strip()

        if choice == "1":
            print("\n--- Current Network Topology Status ---")
            for dev in devices.values():
                dev.display_info()
                print("-" * 35)

        elif choice in ["2", "3"]:
            print("\nSelect target component to manipulate:")
            print("1. Living Room Thermostat\n2. Ceiling Chandelier\n3. Front Porch Security Cam")
            target = input("Target device index: ").strip()

            if target in devices:
                if choice == "2":
                    devices[target].turn_on()
                else:
                    # Defensive programming: stop recording before losing power
                    if target == "3" and isinstance(devices[target], SecurityCamera):
                        devices[target].stop_recording()
                    devices[target].turn_off()
            else:
                print("Invalid device selection index.")

        elif choice == "4":
            sensor.read_temperature()

        elif choice == "5":
            print("\n1. Increase Brightness (+10%)\n2. Decrease Brightness (-10%)")
            adj = input("Selection: ").strip()
            if adj == "1":
                light.increase_brightness()
            elif adj == "2":
                light.decrease_brightness()
            else:
                print("Invalid modification command.")

        elif choice == "6":
            print("\n1. Start Live Recording\n2. Stop Live Recording")
            rec_action = input("Selection: ").strip()
            if rec_action == "1":
                camera.start_recording()
            elif rec_action == "2":
                camera.stop_recording()
            else:
                print("Invalid sequence action command.")

        elif choice == "7":
            print("\nShutting down Smart Home Infrastructure Engine. Goodbye.")
            break
        else:
            print("Invalid command string input entry. Select option from range values.")

if __name__ == "__main__":
    main()
```
